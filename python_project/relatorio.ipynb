```python
#Bloco para importar bibliotecas

from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser

import language_tool_python as lt
```

Exemplos:

```python
#Exemplo1: Extração de texto com o relatório contendo cabeçalho e rodapé e sem alteração no PDF.

output_string = StringIO()
with open('C:\\Users\\Riallen\\Documents\\Att\\python_project\\document_exemplo.pdf', 'rb') as in_file:
    parser = PDFParser(in_file)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    for page in PDFPage.create_pages(doc):
        interpreter.process_page(page)

print(output_string.getvalue())
text = output_string.getvalue()

#Aqui foi gerado um arquivo do tipo .txt para visualização do texto gerado nessa extração.
#Olhar arquivo texto_exemplo1.txt
texto_txt = open("texto_exemplo1.txt", 'x', encoding = 'utf-8')
texto_txt.write(text)
texto_txt.close()
```

```python
#Análise de erros sem filtros
tool = lt.LanguageTool('pt-BR')
matches = tool.check(text)
t = len(matches)
#Além da análise de erros, também foi gerado um arquivo .txt para visualização total dos erros
#Olhar arquivo erro1.txt
aperro = open('erro1.txt', 'x',encoding = 'utf-8')
for i in range(0, t):
    erro = 'Erro: ' + str(i + 1)
    print(erro)
    aperro.write(erro)
    print(matches[i])
    aperro.write(str(matches[i]))
print(t)
aperro.close()
```

```python
#Análise de erros com filtro: 'WHITESPACE_RULE' e 'DASH_RULE'
#Além da análise de erros, também foi gerado um arquivo .txt para visualização total dos erros
#Olhar arquivo erro2.txt
t1 = 0
aperro = open('erro2.txt', 'x', encoding = 'utf-8')
for i in range(0, t):
    if not matches[i].ruleId == 'WHITESPACE_RULE' and not matches[i].ruleId == 'DASH_RULE':
        erro = 'Erro: ' + str(t1 + 1)
        print(erro)
        aperro.write(erro)
        print(matches[i])
        aperro.write(str(matches[i]))
        t1 += 1
print(t1)
aperro.close()
```

```python
#Exemplo2: Documento sem cabeçalho, rodapé e alteração no documento PDF.
#Observe que neste exemplo, outro tipo de documento foi utilizado. Olhar document_test.pdf

output_string = StringIO()
with open('C:\\Users\\Riallen\\Documents\\Att\\python_project\\document_exemplo_tratado.pdf', 'rb') as in_file:
    parser = PDFParser(in_file)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    for page in PDFPage.create_pages(doc):
        interpreter.process_page(page)

print(output_string.getvalue())
text1 = output_string.getvalue()

#Aqui foi gerado um arquivo do tipo .txt para visualização do texto gerado nessa extração.
#Olhar arquivo texto_exemplo2.txt
texto_txt = open("texto_exemplo2.txt", 'x')
texto_txt.write(text)
texto_txt.close()
```

```python
#Análise de erros sem filtros
matches1 = tool.check(text1)
t2 = len(matches1)
#Além da análise de erros, também foi gerado um arquivo .txt para visualização total dos erros
#Olhar arquivo erro3.txt
aperro = open('erro3.txt', 'x',encoding='utf-8')
for i in range(0, t2):
    erro = 'Erro: ' + str(i + 1)
    print(erro)
    aperro.write(erro)
    print(matches1[i])
    aperro.write(str(matches1[i]))
print(t2)
aperro.close()
```

```python
#Análise de erros com filtro: 'WHITESPACE_RULE' e 'DASH_RULE'
#Além da análise de erros, também foi gerado um arquivo .txt para visualização total dos erros
#Olhar arquivo erro4.txt
t3 = 0
aperro = open('erro4.txt', 'x',encoding='utf-8')
for i in range(0, t2):
    if not matches1[i].ruleId == 'WHITESPACE_RULE' and not matches1[i].ruleId == 'DASH_RULE':
        erro = 'Erro: ' + str(t3 + 1)
        print(erro)
        aperro.write(erro)
        print(matches1[i])
        aperro.write(str(matches1[i]))
        t3 += 3
print(t3)
aperro.close()
```
